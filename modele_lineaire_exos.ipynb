```python
import pandas as pd

path = "price_availability.csv"
df_prices = pd.read_csv(path, sep=";")
df_prices.head()
```

```python
path2 = "listings_final.csv"
df_listing = pd.read_csv(path2, sep=";")
df_listing.head()
```

```python
df_prices["local_price"].isna().sum()
```

```python
del df_listing["Unnamed: 0"]
df_listing
```

```python
df_listing.groupby(['listing_id']).mean()
df_merged = pd.merge(df_listing, df_prices, on='listing_id')
df_merged
```

```python
mean_prices = df_merged.groupby(['listing_id'])["local_price"].mean()
```

```python
df_prices = pd.merge(df_prices, mean_prices, on='listing_id')
```
