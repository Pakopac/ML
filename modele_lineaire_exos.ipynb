```python
import pandas as pd

path = "price_availability.csv"
df_prices = pd.read_csv(path, sep=";")
df_prices.head()
```

```python
path2 = "listings_final.csv"
df_listing = pd.read_csv(path2, sep=";")
df_listing.head()
```

```python
df_prices["local_price"].isna().sum()
```

```python
del df_listing["Unnamed: 0"]
#df_listing
```

```python
df_listing.groupby(['listing_id']).mean()
df_merged = pd.merge(df_listing, df_prices, on='listing_id')
#df_merged
```

```python
mean_prices = df_merged.groupby(['listing_id'])["local_price"].mean()
```

```python
df_prices = pd.merge(df_prices, mean_prices, on='listing_id')
```

```python
df = pd.merge(df_listing, df_prices, on='listing_id')
df
```

```python
from sklearn.model_selection import train_test_split
x = df[["person_capacity", "beds","bedrooms", "bathrooms"]]
y = df["local_price_y"]
X_train, X_test, y_train, y_test = train_test_split(x, y, random_state=42,test_size=0.30)
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
from sklearn import linear_model
lm = linear_model.LinearRegression()
model = lm.fit(X_train, y_train)
model
```

```python
model.score(X_test, y_test)
```

```python
model.coef_,model.intercept_
```

```python
prediction = model.predict(X_test)
prediction
```

```python
from sklearn.metrics import mean_squared_error 
mean_squared_error(y_test, prediction)
```
